```python
# ANN Architecture
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import relu,sigmoid
```

```python
ann=Sequential()
```

```python
from keras.api._v2.keras import activations
# build the architecture
ann.add(Dense(units=128,kernel_initializer='he_uniform',activation='relu',input_dim=X_train_data.shape[1]))# 30 inputs given to
ann.add(Dense(units=64,kernel_initializer='he_uniform',activation='relu'))#hidden layer 1 to given hidden layer 2
ann.add(Dense(units=32,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dense(units=16,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dense(units=8,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dense(units=4,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))# hidden layer 2 to output layer
```

```python
# data has given to architecture
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```
